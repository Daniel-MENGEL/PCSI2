## Objectifs et principe

Rappeler l'objectif du TP et la relation du réseau utilisée en incidence normale et à l'ordre 1 :

$$\sin i' = \frac{\lambda}{a}.$$

À compléter : rôle de $\alpha_0$, de $\alpha_1$, du pas $a$ du réseau et des lectures au vernier.


```python
import numpy as np
import matplotlib.pyplot as plt

N = 50_000

def deg2rad(angle):
    return angle*np.pi/180

def rad2deg(angle):
    return angle*180/np.pi

def degmin2deg(deg, minutes):
    return deg + minutes/60

def degmin2rad(deg, minutes):
    return deg2rad(degmin2deg(deg, minutes))
```

## 1. Réglages du goniomètre à réseau

Décrire les réglages optiques, le réglage de l'incidence normale et la lecture de la direction directe $\alpha_0$.


```python
# Direction directe alpha0, lue au vernier
alpha0_deg = None
alpha0_min = None
ualpha0_min = None

assert None not in [alpha0_deg, alpha0_min, ualpha0_min], "Compléter alpha0."

alpha0 = degmin2deg(alpha0_deg, alpha0_min)
ualpha0 = degmin2deg(0, ualpha0_min)

malpha0 = np.random.normal(alpha0, ualpha0, N)

print("Direction directe alpha0 =", malpha0.mean(), "°")
print("Incertitude type :", malpha0.std(ddof=1), "°")
```

## 2. Doublet jaune du sodium

On mesure les directions angulaires des deux raies jaunes à l'ordre $p=1$, puis :

$$\lambda = a\sin |\alpha-\alpha_0|.$$

Pour un réseau de 600 traits par mm, $a=1/600$ mm.


```python
# Lectures au vernier pour les deux raies du doublet jaune du sodium
alpha1_deg = None
alpha1_min = None
ualpha1_min = None

alpha2_deg = None
alpha2_min = None
ualpha2_min = None

assert None not in [alpha1_deg, alpha1_min, ualpha1_min, alpha2_deg, alpha2_min, ualpha2_min], "Compléter les deux lectures du doublet."

alpha1 = degmin2deg(alpha1_deg, alpha1_min)
ualpha1 = degmin2deg(0, ualpha1_min)
alpha2 = degmin2deg(alpha2_deg, alpha2_min)
ualpha2 = degmin2deg(0, ualpha2_min)

malpha1 = np.random.normal(alpha1, ualpha1, N)
malpha2 = np.random.normal(alpha2, ualpha2, N)

a = 1e-3 / 600  # pas du réseau en m
lambda1 = a * np.sin(abs(deg2rad(malpha1 - malpha0))) * 1e9
lambda2 = a * np.sin(abs(deg2rad(malpha2 - malpha0))) * 1e9

print("lambda1 =", lambda1.mean(), "nm ; u(lambda1) =", lambda1.std(ddof=1), "nm")
print("lambda2 =", lambda2.mean(), "nm ; u(lambda2) =", lambda2.std(ddof=1), "nm")
```

```python
# Comparaison aux valeurs attendues du doublet du sodium
lambda1t = 589.0
lambda2t = 589.6

En1 = abs(lambda1t - lambda1.mean()) / lambda1.std(ddof=1)
En2 = abs(lambda2t - lambda2.mean()) / lambda2.std(ddof=1)

print("Écart normalisé pour lambda1 :", En1)
print("Écart normalisé pour lambda2 :", En2)
```

## 3. Vérification de la relation du réseau avec le mercure

On mesure plusieurs raies du spectre du mercure à l'ordre 1. On trace $\sin i'$ en fonction de $\lambda$.

Si l'incidence est bien normale :

$$\sin i'=\frac{\lambda}{a}.$$

La pente attendue vaut donc $1/a$ si $\lambda$ est exprimée dans la même unité que $a$.


```python
# Direction directe avec la lampe Hg, lue au vernier
alpha0_Hg_deg = None
alpha0_Hg_min = None
assert None not in [alpha0_Hg_deg, alpha0_Hg_min], "Compléter alpha0_Hg."
alpha0_Hg = degmin2rad(alpha0_Hg_deg, alpha0_Hg_min)

# Longueurs d'onde du spectre d'émission du mercure Hg, en nm.
# Garder uniquement les longueurs d'onde effectivement mesurées.
lambda_Hg = np.array([404.7, 407.8, 435.8, 491.6, 546.1, 577.0, 579.0, 623.4, 690.7], dtype=float)

# Remplacer chaque couple (deg, min) par vos lectures au vernier.
liste_alpha1 = [
    (None, None),
    (None, None),
    (None, None),
    (None, None),
    (None, None),
    (None, None),
    (None, None),
    (None, None),
    (None, None),
]

assert len(lambda_Hg) == len(liste_alpha1), "Il faut autant de lectures que de longueurs d'onde."
assert all(deg is not None and minute is not None for deg, minute in liste_alpha1), "Compléter toutes les lectures Hg ou retirer les raies non mesurées."

alpha1_Hg = np.zeros_like(lambda_Hg, dtype=float)
for i in range(len(liste_alpha1)):
    alpha1_Hg[i] = degmin2rad(liste_alpha1[i][0], liste_alpha1[i][1])

ip = abs(alpha1_Hg - alpha0_Hg)
sinip = np.sin(ip)

pente, sini = np.polyfit(lambda_Hg, sinip, 1)

plt.figure(figsize=(8, 5))
plt.plot(lambda_Hg, sinip, 'bo', label='Mesures Hg')
plt.plot(lambda_Hg, pente*lambda_Hg + sini, 'r', label='Régression linéaire')
plt.xlabel(r'$\lambda$ en nm')
plt.ylabel(r"$\sin(i')$")
plt.title("Vérification de la relation du réseau")
plt.legend()
plt.show()

print("Pente =", pente, "nm^-1")
print("Ordonnée à l'origine =", sini)
print("Incidence estimée : i =", rad2deg(np.arcsin(sini)), "°")
```

```python
# Valeur expérimentale du pas du réseau à partir de chaque raie
# Comme lambda_Hg est en nm, a_exp est en nm.
a_exp = lambda_Hg / sinip
n_exp = 1e6 / a_exp  # traits par mm

print("Pas expérimental a =", a_exp.mean(), "+/-", a_exp.std(ddof=1), "nm")
print("Nombre de traits par mm =", n_exp.mean(), "+/-", n_exp.std(ddof=1))

a_constructeur = 1e6 / 600  # nm
n_constructeur = 600

E_na = abs(a_exp.mean() - a_constructeur) / a_exp.std(ddof=1)
E_nn = abs(n_exp.mean() - n_constructeur) / n_exp.std(ddof=1)

print("Écart normalisé sur a :", E_na)
print("Écart normalisé sur le nombre de traits par mm :", E_nn)
```

```python
# Grille d'évaluation interactive à compléter par le professeur
# Si les boutons s'affichent, cliquer sur les scores : la note se recalcule automatiquement.
# Si les boutons ne sont pas disponibles, utiliser la partie "saisie numérique" affichée en bas.

def note_grille(objectifs, problematique, protocole, presentation, interpretation, questions):
    return 20 * (0.9 + objectifs/10) \
              * (0.7 + problematique/10) \
              * (0.7 + protocole/10) \
              * (0.7 + presentation/10) \
              * (0.7 + interpretation/10) \
              * (0.7 + questions/10)

try:
    import ipywidgets as widgets
    from IPython.display import display, Markdown, clear_output

    criteres = [
        ("Objectifs", 1),
        ("Problématique", 3),
        ("Protocole", 3),
        ("Présentation", 3),
        ("Interprétation", 3),
        ("Questions résolues", 3),
    ]
    boutons = []
    lignes = []
    for nom, maximum in criteres:
        bouton = widgets.ToggleButtons(
            options=[(str(i), i) for i in range(maximum + 1)],
            value=0,
            description=nom,
            disabled=False,
            button_style="",
            layout=widgets.Layout(width="100%"),
            style={"description_width": "160px"},
        )
        boutons.append(bouton)
        lignes.append(bouton)
    sortie = widgets.Output()

    def afficher_note(*args):
        note = note_grille(*(bouton.value for bouton in boutons))
        with sortie:
            clear_output(wait=True)
            display(Markdown(f"**Note calculée : {note:.1f} / 20**  \n\n**Note arrondie : {round(note)} / 20**"))

    for bouton in boutons:
        bouton.observe(afficher_note, names="value")
    display(widgets.VBox(lignes + [sortie]))
    afficher_note()

except ModuleNotFoundError:
    print("Les boutons interactifs ne sont pas disponibles dans cet environnement.")
    print("Utiliser alors la saisie numérique ci-dessous, puis réexécuter la cellule.")
    objectifs = 0
    problematique = 0
    protocole = 0
    presentation = 0
    interpretation = 0
    questions = 0
    note = note_grille(objectifs, problematique, protocole, presentation, interpretation, questions)
    print("Note calculée :", note, "/ 20")
    print("Note arrondie :", round(note), "/ 20")
```
